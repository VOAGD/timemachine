```python
%matplotlib inline

from qiskit import *
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.visualization import plot_histogram
from qiskit.visualization import plot_state_city, plot_bloch_multivector
from qiskit import Aer, execute
import math 
import numpy as np
```

```python
q = QuantumRegister(2)
c = ClassicalRegister(2)
qc = QuantumCircuit(q, c)
```

```python
qc.reset(0)
qc.reset(1)

qc.h(0)
qc.cx(0, 1)
qc.u((np.pi/2), (np.pi/2), (np.pi/2), 0)
qc.u((np.pi/2), (np.pi/2), (np.pi/2), 1)
qc.name = "Forwards"

qc.measure(0, 0)
qc.measure(1, 1)

qc.draw()
```

```python
backend= Aer.get_backend('statevector_simulator')
job= execute(qc, backend,shots=1)
result = job.result()
```

```python
plot_state_city(result.get_statevector(),color='01')
```

```python
job.result().get_statevector()
```

```python
qq = QuantumCircuit(q, c)

qq.initialize(job.result().get_statevector(),[0,1])
qq.tdg(0) 
qq.tdg(1)
qq.name = "Backwards"

qq.measure(0, 0)
qq.measure(1, 1)
qq.draw()
```

```python
counts  = result.get_counts(qc)
print(counts) 

plot_histogram(counts,
               color=['midnightblue'], title= "Forwards", bar_labels= False)
```

```python
job1 = execute(qq, backend, shots=1000)
result1 = job1.result()
counts1  = result1.get_counts(qq)
print(counts1)

plot_histogram(counts1,
               color=['pink'], title= "Backwards", bar_labels= False)
```

```python
backend = BasicAer.get_backend('statevector_simulator')
result = execute(qc, backend).result()
path = execute(qq, backend).result()
psi  = result.get_statevector(qc)
theta  = path.get_statevector(qq)
```
